```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Vendas_Trimestre.csv', sep = ";", encoding="utf-8")
print(df)
```

```python
df["Valor"] = df["Valor"].str.replace(",", ".").astype(float)

plt.figure(figsize=(12, 6))
plt.bar(df['Trimestre'], df['Valor'])
plt.title('Vendas por trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Valor')
plt.show()
```
